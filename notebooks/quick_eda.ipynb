```python
# Quick EDA + dataset save (Colab / Jupyter)
import pandas as pd
import matplotlib.pyplot as plt

# 1) Load dataset (public URL)
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"
df = pd.read_csv(url)

# 2) Quick checks
print("Shape:", df.shape)
display(df.head())
print("\nMissing values:\n", df.isnull().sum())
display(df.describe(include='all'))

# 3) Add a derived feature: tip percentage
df['tip_pct'] = (df['tip'] / df['total_bill']) * 100

# 4) Basic plots (inline)
plt.figure(figsize=(8,4))
plt.hist(df['total_bill'], bins=25)
plt.title('Distribution of total_bill')
plt.xlabel('total_bill')
plt.ylabel('count')
plt.show()

plt.figure(figsize=(8,4))
df.groupby('day')['tip_pct'].mean().sort_values(ascending=False).plot(kind='bar')
plt.title('Average tip % by day')
plt.ylabel('avg tip %')
plt.show()

# 5) Save cleaned dataset to disk (Colab: /content)
out_path = '/content/tips_cleaned.csv'
df.to_csv(out_path, index=False)
print("Saved cleaned CSV to:", out_path)

# 6) (Optional) display top rows again
display(df.head())
```
